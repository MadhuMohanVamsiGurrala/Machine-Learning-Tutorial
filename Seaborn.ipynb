```python
#import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import seaborn
import seaborn as sns
```

```python
#import dataset
tips = sns.load_dataset('tips')
```

```python
print(tips.head())
```

```python
warnings.simplefilter(action = "ignore", category = UserWarning)
print(sns.get_dataset_names())
```

Plotting Univariate Distribution with distplot()

```python
sns.distplot(tips['total_bill'])
plt.show()
```

Seaborn Histogram

```python
sns.distplot(tips['total_bill'], kde=False)
plt.show()
```

Seaborn - Kernel density estimates

```python
sns.distplot(tips['total_bill'], hist=False)
plt.show()
```

Plotting Bivariate Distribution with jointplot()

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips)
plt.show()
```

Seaborn - Scatter Plot

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
plt.show()
```

Visualizing pairwise relationship with pairplot()

```python
sns.set_style("ticks")
sns.pairplot(tips)
plt.show()
```

seaborn - strip plot

```python
sns.stripplot(x = "day", y = "total_bill", data = tips)
plt.show()
```

```python
sns.stripplot(x ="day", y = "total_bill", data = tips, jitter = True)
plt.show()
```

Seaborn - Swarm plot

```python
sns.swarmplot(x = "day", y = "total_bill", data = tips)
plt.show()
```

Seaborn - Boxplot

```python
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
sns.boxplot(x = "day", y = "total_bill", data = tips)
plt.show()
```

Seaborn - Violin Plot

```python
sns.violinplot(x = tips["total_bill"])
plt.show()
```

Seaborn - Bar Plot

```python
sns.barplot(x= "day", y = "total_bill", data = tips)
plt.show()
```

Seaborn - Point Plot

```python
sns.pointplot(x = "time", y = "total_bill",  data = tips)
plt.show()
```

```python
sns.pointplot(x= "time", y = "total_bill", hue = "smoker", data = tips)
plt.show()
```

seaborn - regplot

```python
sns.regplot(x = "total_bill", y = "tip", data = tips)
plt.show()
```

Seaborn - Heat map

```python
uniform_data = np.random.rand(10,12)
sns.heatmap(uniform_data)
plt.show()
```
