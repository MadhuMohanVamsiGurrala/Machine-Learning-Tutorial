```python
import pandas as pd
```

```python
# Generate 50 country names
countries = ['Country' + str(i) for i in range(1, 51)]

# Create data for 50 countries
data = {
    'COUNTRY': countries,
    'POP': [i * 1000000 for i in range(1, 51)],
    'AREA': [i * 10000 for i in range(1, 51)],
    'GDP': [i * 100000 for i in range(1, 51)],
    'CONT': ['Continent' + str((i % 5) + 1) for i in range(1, 51)],
    'IND_DAY': ['Day' + str(i) for i in range(1, 51)]
}
```

```python
data
```

```python
df = pd.DataFrame(data = data).transpose()
df
```

```python
df.to_csv('data.csv') #saving data as csv file
```
